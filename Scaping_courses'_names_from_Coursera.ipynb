```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import time

#use BeautifulSoup for easier html extraction
from bs4 import BeautifulSoup
import urllib.request
```

```python
#functions:

#1) used for identify specific element at the page using
#   ID ot CLASS befor continue.
delay=3
def wait_until_elem(type,selec,delay):

    try:
        wait = WebDriverWait(driver, delay)
        if type=='ID':
            wait.until(EC.presence_of_element_located((By.ID, selec))) 
        elif type=='CLASS':
            wait.until(EC.presence_of_element_located((By.CLASS_NAME, selec))) 
        #wait until searchform loadded to the page
        #we search here using the ID
        print("Page is ready")
    except TimeoutException:
        print("Loading took too much time")
        
def datascience_open():
    #open first category
    wait_until_elem('ID',"rendered-content",3)
    driver.find_element_by_xpath('//div[@data-index="0"]').click()

#check page:
    wait_until_elem("CLASS","filter-bar-placeholder",3)   
    
#Function: click on skill button
def skill_Button_CLK():
    skills = driver.find_elements_by_class_name('Button_1w8tm98-o_O-icon_1rbfoc-o_O-sm_pd780u')[0]
    skills.click()
    wait_until_elem('CLASS',"content_3x0z5j-o_O-Box_120drhm-o_O-displayflex_poyjc-o_O-columnDirection_ia4371",3)
    
    
#Scrap the names of all the skills:
def skills_scraping():
    skills_names=[]
    skills=driver.find_elements_by_class_name('input_3xha1c-o_O-active_ro0g1e') #this class is const for all the skills
    for i in range(len(skills)):
        skills_names.append(skills[i].get_attribute('value'))
    return skills_names    

#click skills
#with the change of the number, we click on the skill of that index
def skill_click(i):
    driver.find_elements_by_class_name('input_3xha1c-o_O-active_ro0g1e')[i].click()
    

#click on apply button
def apply_clk():
    driver.find_elements_by_class_name('Button_1w8tm98-o_O-primary_cv02ee-o_O-md_1jvotax')[0].click()    
    
#scrap the names of all the courses:
def courses_scraping():
    #scrap courses names:
    courses_names=[]
    #time.sleep(1) 
 
    wait_until_elem("CLASS","sortContainer_7balsc",10)
    #get soup
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    courses_section = soup.find("div",{"class":"ReactVirtualized__Grid__innerScrollContainer"})
    Courses = courses_section.find_all("a", {"data-click-key":"browse.browse.click.offering_card"})

    for i in range(len(Courses)):
        courses_names.append(Courses[i]['aria-label'])
    return courses_names

#clear the filters
def clear_filters_CLK():
    driver.find_elements_by_class_name('Button_1w8tm98-o_O-link_wjki9h-o_O-zero_6k3h32')[0].click()
```

```python
#open coursera page
driver = webdriver.Chrome()
driver.implicitly_wait(2)
driver.get("https://www.coursera.org/browse")
wait_until_elem('ID',"rendered-content",3)
datascience_open()
```

```python
#open "datascience actegory":
skill_Button_CLK()
skills_names=skills_scraping()
```

```python
courses={}
for i in range(len(skills_names)):
    skill_Button_CLK()
    skill_click(i)
    apply_clk()
    courses_names=courses_scraping()
    courses[skills_names[i]]=courses_names
    clear_filters_CLK()
    print(i)
```
